```python
number = int(input("Enter a number: "))

if number>0:
    print("Positive")

elif number<0:
    print("Negative")

else:
    print("Zero")
```

```python
marks = int(input("Enter student marks : "))
if marks>=90:
    print("grade a ")
elif marks>=75:
    print("grade b ")
elif marks>=60:
    print("grade c")
else:
    print("fail")
```
