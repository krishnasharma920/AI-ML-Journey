```python
 def add(a,b):
     return a + b
 def sub(a,b):
    return a - b 
 def mult(a,b):
     return a*b
 def divide(a,b):
    return a/b
 def square(n):
     return n*n
 def cube(n):
     return n*n*n

while True:
    print("\n--- Calculator Menu ---")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Square")
    print("6. Cube")
    print("7. Exit")
    choice = input("Enter you choice ")

    if choice in ("1","2","3","4"):
        m = float(input("Enter first numbeer:  "))
        n = float(input("Enter second number:  "))

        if choice == "1":
            print("Resut = ", add(m,n))
        elif choice == "2":
            print("result = ", sub(m,n))
        elif choice == "3":
            print("result = ", mult(m,n))
        elif choice == "4":
            print("result = ", divide(m,n))
    elif choice == "5":
        n = float(input("Enter a number: "))
        print("result = ", square(n))
    elif choice == "6":
        n = float(input("Enter a number:  "))
        print("result = ", cube(n))
    else:
        print("Invalid number , please choose a valid number:")
    choice = input("Do you want to continue: (yes/no)")
    if choice.lower() != "yes":
       break

    

    
     
```
